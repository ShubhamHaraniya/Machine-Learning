```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(train_data['prognosis'])
train_data['prognosis'] = le.transform(train_data['prognosis'])
```

```python
train_data[train_data['prognosis']==4]
```

```python
lab_map = dict(zip(le.transform(le.classes_),le.classes_))
lab_map
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
models = {
    'RFC' : RandomForestClassifier(n_estimators=100,random_state=42),
    'GBC' : GradientBoostingClassifier(n_estimators=100,random_state=42),
    'MNB' : MultinomialNB(),
    'SVC' : SVC(kernel='linear')
}
```

```python
X = train_data.drop('prognosis', axis=1)
y = train_data['prognosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.2,random_state=42)
```

```python
for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = model.predict(X_test)
    
    # Calculate accuracy
    accuracy = accuracy_score(y_true=y_test, y_pred=y_pred)
    
    # Print the name of the model and its accuracy
    print(f"{name} : {accuracy}")
```

```python
# Select SVC as Model

model = SVC(kernel='linear')
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

accuracy_score(y_true=y_test,y_pred=y_pred)
```
